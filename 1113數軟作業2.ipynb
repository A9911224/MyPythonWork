## 圓與線性變換

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
t=np.linspace(0,2*np.pi,100)
x=np.cos(t)
y=np.sin(t)
plt.plot(x,y)
plt.axis('equal')
plt.show()
```

```python
a=np.random.randn(2,2)
b=np.array((x,y))
np.dot(a,b)
plt.plot(np.dot(a,b)[0],np.dot(a,b)[1])
plt.axis('equal')
plt.show()
```
