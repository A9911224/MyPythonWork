```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### 非線性回歸與最小平方法

```python
num = 70
```

```python
x = np.linspace(0,1000,num = num)
y = np.sin(x) + (1/20)*np.cos(2*x) + 2*np.random.randn(num)
plt.plot(x,y)
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.sin(2*x)
y4 = np.cos(2*x)
y5 = np.sin(3*x)
y6 = np.cos(3*x)
y7 = np.sin(4*x)
y8 = np.cos(4*x)
y9 = np.sin(5*x)
y10 = np.cos(5*x)
```

```python
A = np.stack((y1, y2, y3, y4, y5, y6, y7, y8, y9, y10, np.ones(num)),axis = 1)
```

```python
A.shape
```

```python
np.linalg.inv(A.T@A)@A.T@y
```

```python
def linear_regression(t):
     return (np.linalg.inv(A.T@A)@A.T@y)@np.array([np.sin(t), np.cos(t), np.sin(2*t), np.cos(2*t), np.sin(3*t),np.cos(3*t), np.sin(4*t), np.cos(4*t), np.sin(5*t), np.cos(5*t), 1])
```

```python
y_pre = [linear_regression(t) for t in x] 
```

```python
plt.plot(x,y_pre)
plt.show()
```

```python
plt.plot(x,y, 'r' ,label = "obs")
plt.plot(x,y_pre , 'b', label = 'pred')
plt.legend()
plt.show()
```
