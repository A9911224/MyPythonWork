```python
##印出n*n的菱形 (n須為偶數)
a = int(input())
if a%2 ==0:
    print("you cannot print")
if a<= 0:
    print("you cannot print")
if a%2 ==1:
    for i in range(1,int((a+1)/2)):
        print(int((a-1-2*i)/2)*" ",((i*2-1))*"*")
    print(a*"*")
    for j in range(1,int((a+1)/2)):
        print((j-1)*" ",((a-2*j))*"*")
```
